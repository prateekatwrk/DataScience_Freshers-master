## map and filter

```python
seq = [1,2,3,4,5]
```

```python
list(map(lambda var: var*2,seq))
```

```python
list(filter(lambda item: item%2 == 0,seq))
```

## reduce

```python
from functools import reduce
reduce((lambda x, y: x + y), seq)
```

```python
## Class

**Every class must have a constructor, even if you don’t create a constructor explicitly it will create a default constructor by itself.**
```

```python
#Creating a class
class Person:
    x='Hi'
    #using  __init__() function to assign values for name and age
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
#creating objects        
p1 = Person("Ram", 30)

print(p1.x)
print(p1.name)
print(p1.age)
```

**The self parameter is a reference to the current instance of the class and is used to access variables that belongs to the class. It has to be the first parameter of any function but not neccesarily named self.**

## Inheritance

```python
class Operations:
    a = 10
    b = 20
    
    def add(self):
        sum = self.a + self.b
        print("Addition of a and b is :", sum)
        
class Add(Operations):
    c = 50
    d = 10
    def sub(self):
        sub = self.c +self.d
        print("Subtraction of c and d is: ", sub)
        
ob = Add()
ob.add()
ob.sub()
```

## Polymorphism

**Method overriding: If a subclass method has the same name which is declared in the superclass method then it is called Method overriding**

```python
class Animal:
    def breathe(self):
        print("I breathe oxygen")
        
    def feed(self):
        print("I eat food")
        
class Herbivorous(Animal):
    def feed(self):
        print("I eat only plants")
        
ob = Herbivorous()
ob.feed()
ob.breathe()
```
