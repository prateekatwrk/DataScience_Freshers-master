## Line Plot: 
This is the simplest of all plotting type, as it is the visualization of a single function.

```python
#world population graph
years=[1,1000,1500,1600,1700,1750,1800,1850,1900,1950,1955,1960,1965,1970,1980,1985,1990,
       1995,2000,2005,2010,2015]
pops=[200,400,458,580,682,791,1000,1262,1650,2525,2758,3018,3322,3682,
      4061,4440,4853,5310,5735,6127,6520,7349]

plt.plot(years,pops)
plt.show()
```

```python
years=[1,1000,1500,1600,1700,1750,1800,1850,1900,1950,1955,1960,1965,1970,1980,1985,1990,
       1995,2000,2005,2010,2015]
pops=[200,400,458,580,682,791,1000,1262,1650,2525,2758,3018,3322,3682,
      4061,4440,4853,5310,5735,6127,6520,7349]

plt.plot(years,pops)
plt.xlabel('years')
plt.ylabel('Population')
plt.title('Population growth')
plt.show()
```

```python
x = [1,2,3]
y = [4,5,6]
x2 = [2,3,4]
y2 = [6,7,8]

#label is used to give a label to each line
plt.plot(x, y, label = 'First Line', color='red', linestyle='solid')
plt.plot(x2, y2, label = 'Second Line', color='g', linestyle='dashed')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Graph')
#legend is used to plot legend that labels each line type when multiple lines are being shown.
plt.legend()
plt.show()
```

## Bar Chart: 
A Bar Graph is used to compare data among different categories. Bar Chart can be represented Horizontally and Vertically.

```python
x = [2,4,6, 8, 10]
y = [4,2,5,6,7]

#vertical bar chart; for horizontal bar chart use barh
plt.bar(x,y, label = 'Bar1')

plt.xlabel('x-axis')
plt.title('Bar Graph')
plt.legend()
plt.show()
```

## Histograms: 
Histograms are similar to Bar Chart, however, Histograms are used to show distribution. These are helpful when we have data as arrays.

Examples of this might be age groups, or scores on a test.

```python
#using numpy library to draw random samples from a normal distribution.
normal_data = np.random.normal(size= 1000)

#hist() is used to plot histograms
plt.hist(normal_data, bins = 10)

plt.xlabel('Normalized data')
plt.ylabel('Frequency')
plt.title('Histogram Graph')
plt.show()
```

**Bin refers to the range of values that are divided into a series of usually the same size of intervals.
in the output we can see that value between 0 and 1 has the highest frequency.**

## Scatterplot
Scatter Plot is much alike to line graph in which instead of points being joined by line segments, the points are shown individually with a dot, circle or any other shape.

Scatter plot shows the relationship between two variables

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]

plt.scatter(x, y, color='black', marker = 'o')
plt.show()
```

## Pie Chart
A pie chart is a circular statistical diagram. The area of the whole chart represents the whole of the data. The areas of the pie chart represents the percentage of parts of data and it is called wedges.

```python
day = [1,2,3,4,5]
sleeping = [7, 8, 6, 11, 7]
eating = [2, 3, 4, 3, 2]
working = [7, 8, 7, 2, 2]
playing = [8, 5, 7, 8, 13]

slices = [7,2,7,8]
activities = ['sleeping', 'eating', 'working', 'playing']
cols = ['lightgreen', 'lightblue', 'lightcoral', 'cyan']

#pie() method is used to plot pie chart
plt.pie(slices, labels = activities, colors = cols, startangle = 90, shadow = True, explode = (0, 0.1, 0, 0))
plt.show()
```

## Stack Plot
A stack plot is a plot which shows the whole data set with easy visualization of how each part makes up the whole.
Each constituent of the stack plot is stacked on top of each other.

```python
days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating =   [2,3,4,3,2]
working =  [7,8,7,2,2]
playing =  [8,5,7,8,13]

plt.stackplot(days, sleeping,eating,working,playing, colors=['plum','lightgreen','c','salmon'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nStack Plots')
plt.show()
```

```python
days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating =   [2,3,4,3,2]
working =  [7,8,7,2,2]
playing =  [8,5,7,8,13]

plt.plot([],[],color='plum',      label='Sleeping', linewidth=5)
plt.plot([],[],color='lightgreen',label='Eating', linewidth=5)
plt.plot([],[],color='c',       label='Working', linewidth=5)
plt.plot([],[],color='salmon', label='Playing', linewidth=5)

plt.stackplot(days, sleeping,eating,working,playing, colors=['plum','lightgreen','c','salmon'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Stack Plot')
plt.legend()
plt.show()
```

## barplot

```python
sns.barplot(x = 'sex', y = 'total_bill', data = df)
```

**The black bar in the above graph shows the uncertainty or the estimate of standard error that is occurence of the data is less frequent. The error bars are used to display either the Standard deviation, Standard error, Confidence interval and maximum & minimum values in a ranged dataset.
The short error bars shows that the values are concentrated signalling that the plotted average value is most likely while the longer error bars shows that the values are more spread out and less reliable.
However if the data is skewed then the length on each side would be unbalanced.**

```python
#figure size width, height in inches; default value (6.4) & (4.8)
#plt.figure(figsize=(7,7))
sns.barplot(x = df['smoker'].value_counts().index, y = df['smoker'].value_counts().values, data = df)
plt.xlabel('smoker')
plt.ylabel('Frequency')
plt.title('Smoker Bar Plot')
plt.show()
```

## boxplot and violinplot
boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
sns.boxplot(x="sex", y="total_bill", data=df,palette='rainbow')
```

```python
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

## violinplot
A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
sns.violinplot(x="day", y="total_bill", data=df,palette='rainbow')
```
