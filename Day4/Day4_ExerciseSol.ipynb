```python
#import all the neccessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
#load and dataset and retrieve the first few records
df = pd.read_csv(r'C:\Users\DELL\Downloads\StudentsPerformance.csv')
df.head()
```

```python
#understand the dataset and retrieve its information and make inferences
df.info()
```

```python
#Check for null values
df.isnull().sum()
```

```python
#Display gender counts using value_counts() function
df['gender'].value_counts()
```

## Visualization

```python
#Visualize the count of gender
sns.countplot(x = 'gender', data = df)
```

```python
#Draw a barplot showing the race/ethnicity 
sns.barplot(x = df['race/ethnicity'].value_counts().index, y = df['race/ethnicity'].value_counts().values)
plt.xlabel('race/ethnicity')
plt.ylabel('frequency')
plt.title('Race/Ethnicity Bar Plot')
plt.show()
```

```python
sns.countplot(x = 'race/ethnicity', data = df)
```

```python
#plot 3 respective graphs showing the comparison for ‘writing score’, ‘reading score’ and ‘math score’ for both the ‘genders’ based on ‘
#parental level of education’
```

```python
sns.barplot(x = "parental level of education", y = "writing score", hue = "gender", data = df)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(x = "parental level of education", y = "reading score", hue = "gender", data = df)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(x = "parental level of education", y = "math score", hue = "gender", data = df)
plt.xticks(rotation=45)
plt.show()
```

```python
#plot a heatmap and find the corelation between the variables
df1 = df.corr()
sns.heatmap(df1, annot = True)
```

```python
#plot a pairplot for the dataset
sns.pairplot(df)
```

```python
#plot a graph for math score vs writing score for both the genders using lmplot
sns.lmplot(x = 'math score', y = 'writing score', hue = 'gender', data = df)
```

```python
sns.jointplot(x = 'math score', y = 'writing score', data = df)
```

```python
#Visualize for math score for both the genders using Box Plot
sns.boxplot(x = 'gender', y = 'math score', data = df)
```
